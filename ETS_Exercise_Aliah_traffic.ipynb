## ETS Decomposition of Airline Passengers Dataset:

## Import libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
```

## Read dataset

```python
traffic = pd.read_csv('/content/traffic.csv',
                       index_col ='DateTime',
                       parse_dates = True )
# I tried to apply ETS on traffic dataset
```

## show dataset

```python
traffic.sample(10)
```

## Apply ETS

```python
result = seasonal_decompose(traffic['Vehicles'],
                            model ='multiplicative', period=7)
```

## Plot ETS

```python
result.plot()
```
